# Creating Dataset

```python
import requests

url = 'https://www.openml.org/data/download/22102255/dataset'

r = requests.get(url, allow_redirects=True)
```

```python
with open('dataset.txt', 'wb') as f:
  f.write(r.content)
```

```python
import pandas as pd

data = []
with open('dataset.txt', 'r') as f:
  for line in f.read().split('\n'):
    if line.startswith('@') or line.startswith('%') or line == "":
      continue
    data.append(line)
```

```python
columns = []
with open('dataset.txt', 'r') as f:
  for line in f.read().split('\n'):
    if line.startswith('@ATTRIBUTE'):
      columns.append(line.split(' ')[1])
```

```python
columns
```

```python
data
```

```python
with open('df.csv', 'w') as f:
  f.write(','.join(columns))
  f.write('\n')
  f.write('\n'.join(data))
```

```python
df = pd.read_csv('df.csv')
df.columns = columns
```

```python
df
```

```python
df['t_win'] = df.round_winner.astype('category').cat.codes
```

```python
df
```

# Creating Visualizations and Performing Feature Selection

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
correlations = df[columns+['t_win']].corr()
print(correlations['t_win'].apply(abs).sort_values(ascending=False).iloc[:25])
```

```python
selected_cols = []
for col in columns+['t_win']:
  try:
    if abs(correlations[col]['t_win'])>0.15:
      selected_cols.append(col)
  except KeyError:
    pass
df_selected = df[selected_cols]
```

```python
df_selected
```

```python
plt.figure(figsize=(16, 10))
sns.heatmap(df_selected.corr().sort_values(by='t_win'), annot=True, cmap='YlGnBu')
```

```python
df_selected.hist(figsize=(18,12))
plt.show()
```

# Machine Learning

```python
from sklearn.model_selection import train_test_split

X, y = df_selected.drop(['t_win'], axis=1), df_selected['t_win']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=420)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier

scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.fit_transform(X_test)

knn = KNeighborsClassifier()
knn.fit(X_train_scaled, y_train)
```

```python
knn.score(X_test_scaled, y_test)
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
grid_param = {
    'n_neighbors' : list(range(5, 20, 3)),
    'weights' :['uniform', 'distance']
}

knn = KNeighborsClassifier(n_jobs=4)
```

```python
classifier = RandomizedSearchCV(knn, grid_param, n_jobs=4, n_iter=3, verbose=2, cv=3)
classifier.fit(X_train_scaled, y_train)
```

```python
knn = classifier.best_estimator_
```

```python
knn.score(X_test_scaled, y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier

forest = RandomForestClassifier()
forest.fit(X_train_scaled, y_train)
```

```python
forest.score(X_test_scaled, y_test)
```

```python
grid_param = {
    'n_estimators' : list(range(80, 130, 10)),
    'criterion' :['gini', 'entropy']
}

forest = RandomForestClassifier(n_jobs=4)
classifier = RandomizedSearchCV(forest, grid_param, n_jobs=4, n_iter=3, verbose=2, cv=3)
classifier.fit(X_train_scaled, y_train)

forest_optimized = classifier.best_estimator_
forest_optimized.score(X_test_scaled, y_test)
```

**We can see that our optimized KNN gets an accuracy of 78%+ and RandomForest Optimized gets 82%+ accuracy.**
